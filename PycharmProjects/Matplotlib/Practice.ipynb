```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.figure(figsize=(8,4), facecolor='yellow')

plt.plot(x,y,label='2x',color='pink',linestyle='dashdot',linewidth=4,marker='*',markersize='20',markeredgecolor='blue',markerfacecolor='orange')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'g', label='x^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')



plt.title('Our First Graph', fontdict={'fontsize':20,'fontname':'Comic Sans MS'})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks(np.arange(0,20,2))
plt.legend()
plt.savefig('mygraph.png', dpi=1000)
plt.show()
```

```python
cat = ['A','B','C']
num = [1,4,2]
bars = plt.bar(cat,num,color='orange',hatch='+')
patterns = ['*','/','O']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch('O')
# bars[1].set_hatch('o')
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize':18})
plt.plot(gas['Year'], gas['USA'], 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas['Year'], gas['Australia'], 'y.-', label='Australia')
plt.xticks(gas['Year'][::3])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize':18})

for country in gas.columns:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], label=country, marker='*')

plt.xticks(gas['Year'][::3])
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()
plt.savefig('gas_price_figure.png', dpi=1000)
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head(5)
```

```python
bins = [40, 50, 60, 70, 80, 90, 100]
plt.figure(figsize=(8, 8))
plt.hist(fifa['Overall'], bins = bins, color = 'red')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.xlabel('Skill Level')
plt.ylabel('Number of players')
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0] 
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
plt.pie([left,right], labels=['Left foot players','Right foot players'], autopct='%.1f%%')
plt.title("Distribution of Player's preferred foot in FIFA 2018")
plt.show()
```

```python
fifa['Weight']
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa['Weight']]
fifa['Weight']
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

plt.title('Weight distribution of FIFA players (in lb)')
plt.style.use('ggplot')
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.5,0.2,0,0,0.5)
plt.pie(weights, labels=labels, autopct='%.1f%%', explode = explode)
plt.show()
```

```python
plt.style.use('default')

bayern = fifa.loc[fifa['Club']=='FC Bayern München']['Overall']
sounders = fifa.loc[fifa['Club']=='Seattle Sounders FC']['Overall']

boxes = plt.boxplot([bayern, sounders], labels=['FC Bayern München', 'Seattle Sounders FC'], patch_artist=True)

for box in boxes['boxes']:
    box.set(color='blue', linewidth=2)
    box.set(facecolor='yellow')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
